##Simple program to test out different downsampling functions from various Python libraries.

Will be helpful as the Acoustic Species Identification project move s forward as the sampling rate of the Audiomoths has been set to 384kHz, when most research papers we have looked into use 44.1kHz in order to focus on the human hearing range.

Based off of this page: https://nbviewer.jupyter.org/gist/jthiem/0bb8b2296c8dbd0b83bfab8270a8eb24

Jacob Ayers 6/24/2020

```python
#Package Imports
import numpy as np
import matplotlib.pyplot as plot
import scipy
from scipy.io import wavfile

import samplerate
import scipy.signal as scipy_signal
import resampy
```

Plotting the raw audio fle

```python
WAV_FILE = "Audiomoth_Test_Clip"
#Sampling rate and signal of a clip from the deployed Audiomoth Array
BASE_SAMPLING_RATE, SIGNAL = wavfile.read(WAV_FILE+".WAV");
#Human hearing range we want to downsample to
DOWNSAMPLED_RATE = 44100;

#Choosing the resample function to perform well, rather than quickly ("sinc_fastest")
CONVERSION_METHOD = "sinc_best"

RATE_RATIO = DOWNSAMPLED_RATE/BASE_SAMPLING_RATE;
```

```python
#Plotting the raw spectrogram

plot.subplot(7,1,1)
plot.figure(figsize=(15,3));
plot.specgram(SIGNAL,Fs=BASE_SAMPLING_RATE,NFFT = 1024, noverlap = 512, window = np.hanning(1024))
plot.colorbar()
plot.ylabel("Frequency (Hz)")
plot.xlabel("Time (s)")
```

```python
#Plotting the raw spectrogram limited to human hearing range
plot.subplot(7,1,2)
plot.figure(figsize=(15,3));
plot.specgram(SIGNAL,Fs=BASE_SAMPLING_RATE,NFFT = 1024, noverlap = 512, window = np.hanning(1024))
plot.colorbar()
plot.ylabel("Frequency (Hz)")
plot.xlabel("Time (s)")
plot.ylim(0,22050)
```

# Baseline Downsampling with Audacity
"Tracks"==>"Resample"==>44100

```python
plot.subplot(7,1,3)
Audacity_Rate, Audacity_Signal = wavfile.read(WAV_FILE+"_audacity_downsample.WAV");
plot.figure(figsize=(15,3));
plot.specgram(Audacity_Signal,Fs=Audacity_Rate,NFFT = 1024,noverlap = 512, window = np.hanning(1024))
plot.colorbar()
plot.ylabel("Frequency (Hz)")
plot.xlabel("Time (s)")
plot.ylim(0,22050)
```

# Implementing "Secret Rabbit Code" Downsampling method
Better known as libsamplerate in Python 3
Same as scikits.samplerate in Python 2 (I know this since I wasted several hours importing the wrong package)
https://pypi.org/project/samplerate/

```python
#plotting the downsampled spectrogram
plot.subplot(7,1,4)
RESAMPLED_SIGNAL = samplerate.resample(SIGNAL,RATE_RATIO,CONVERSION_METHOD)
plot.figure(figsize=(15,3));
plot.specgram(RESAMPLED_SIGNAL,Fs=DOWNSAMPLED_RATE,NFFT = 1024, noverlap = 512, window = np.hanning(1024))
plot.colorbar()
plot.ylabel("Frequency (Hz)")
plot.xlabel("Time (s)")

#Writing the new downsampled clip to compare via sound.
wavfile.write(WAV_FILE+"_libsamplerate_Downsample.WAV",DOWNSAMPLED_RATE,RESAMPLED_SIGNAL)
#Sounds a lot louder (as evidenced in the spectrogram with the brighter power)
```

# Implementing scipy's built in resample function
https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.resample.html

```python
plot.subplot(7,1,5)
RESAMPLED_SIGNAL = scipy_signal.resample(SIGNAL,int(len(SIGNAL)*RATE_RATIO))
plot.figure(figsize=(15,3));
plot.specgram(RESAMPLED_SIGNAL,Fs=DOWNSAMPLED_RATE,NFFT = 1024, noverlap = 512, window = np.hanning(1024))
plot.colorbar()
plot.ylabel("Frequency (Hz)")
plot.xlabel("Time (s)")
#Writing the new downsampled clip to compare via sound.
wavfile.write(WAV_FILE+"_scipy_resample_Downsample.WAV",DOWNSAMPLED_RATE,RESAMPLED_SIGNAL)
#This also sounds terrible. Makes we think that Matlab is the way to go on this front
```

# Resampy Implementation
https://resampy.readthedocs.io/en/master/example.html

```python
plot.subplot(7,1,6)
RESAMPLED_SIGNAL = resampy.resample(np.float64(SIGNAL),BASE_SAMPLING_RATE,DOWNSAMPLED_RATE)
plot.figure(figsize=(15,3));
plot.specgram(RESAMPLED_SIGNAL,Fs=DOWNSAMPLED_RATE,NFFT = 1024, noverlap = 512, window = np.hanning(1024))
plot.colorbar()
plot.ylabel("Frequency (Hz)")
plot.xlabel("Time (s)")
#Writing the new downsampled clip to compare via sound.
wavfile.write(WAV_FILE+"_resampy_Downsample.WAV",DOWNSAMPLED_RATE,RESAMPLED_SIGNAL)
```

# Scipy Decimate Implementation
https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.decimate.html

```python
plot.subplot(7,1,7)
RESAMPLED_SIGNAL = scipy_signal.decimate(SIGNAL, int(BASE_SAMPLING_RATE/DOWNSAMPLED_RATE))
plot.figure(figsize=(15,3));
plot.specgram(RESAMPLED_SIGNAL,Fs=DOWNSAMPLED_RATE,NFFT = 1024, noverlap = 512, window = np.hanning(1024))
plot.colorbar()
plot.ylabel("Frequency (Hz)")
plot.xlabel("Time (s)")
#Writing the new downsampled clip to compare via sound.
wavfile.write(WAV_FILE+"_decimate_Downsample.WAV",DOWNSAMPLED_RATE,RESAMPLED_SIGNAL)
#The audio quality on this is much better than on the previous downsampling techniques.
```
