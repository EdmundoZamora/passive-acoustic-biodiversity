```python
from microfaune_package.microfaune.detection import RNNDetector
import csv
import os
import glob
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("Stratified_Random_Sample_Bird_Labels.csv")
```

```python
bird_detector = RNNDetector()
for ndx in dataset.index:
    #dataset["Microfaune_Global_Score"][ndx] = 999
    folder_name = dataset["Folder"][ndx]
    clip_name = dataset["Clip"][ndx]
    clip_path = "/home/jacob/Acoustic-Species-Identification/Stratified_Random_Sample_Peru2019_Subset/" + folder_name + '/' + clip_name
    global_score, local_score = bird_detector.predict_on_wav(clip_path)
    print(folder_name + ' ' + clip_name + ' Global Score: ' + str(global_score))
    dataset["Microfaune_Global_Score"][ndx] = global_score
dataset.to_csv("Stratified_Random_Sample_Microfaune_Global_Scores.csv")
```
